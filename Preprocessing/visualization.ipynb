```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_prices_returns(stock, year):
    prices_df = pd.read_csv(f'/Users/raunaksandhu/Desktop/UNI/4th YR/Winter/MATH 518/FTS_Forecasting/ProcessedCSVData/{year}_{stock}_prices.csv')
    plt.plot(prices_df['TimeStamp'], prices_df[stock])
    plt.title(f'{stock} Prices in {year}')
    plt.xlabel('Time')
    plt.ylabel('Price')
    plt.show()

    returns_df = pd.read_csv(f'/Users/raunaksandhu/Desktop/UNI/4th YR/Winter/MATH 518/FTS_Forecasting/ProcessedCSVData/{year}_{stock}_returns.csv')
    plt.plot(returns_df['TimeStamp'], returns_df['returns'])
    plt.title(f'{stock} Returns in {year}')
    plt.xlabel('Time')
    plt.ylabel('Daily Returns')
    plt.show()
```

```python
plot_prices_returns('SPY', 2020)
plot_prices_returns('QQQ', 2020)

# note the stock split in 2020
plot_prices_returns('AAPL', 2020)

plot_prices_returns('MSFT', 2020)
```
